```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('fake_bills(1).csv')
```

```python
data
```

```python
# first we gotta clean teh data
```

```python
data.values
```

```python
rows = []

for i in range(data.shape[0]):
    rows.append(data.loc[i].str.split(';').values)
```

```python
rows
```

```python
cols = data.columns.str.split(';')
cols
```

```python
def colInfo(n):
    info = []
    for i in range(len(rows)):
        info.append(rows[i][0][n])
    return info
```

```python
col1 = colInfo(0)
col2 = colInfo(1)

col3 = colInfo(2)
col4 = colInfo(3)
col5 = colInfo(4)
col6 = colInfo(5)
col7 = colInfo(6)
```

```python
data =  pd.DataFrame({ "is_genuine":col1,
              "diagonal": col2,
              "height_left":col3,
              "height_right": col4,
              "margin_low":col5,
              "margin_up":col6,
              "length":col7
})
```

```python
data
```

# True = 0 and False =1

```python
data.info()
```

```python
def convert(n):
    """Converts a number to a string and then to a list of characters."""
    n = str(n)
    if n =='True':
        return 0.0
    else:
        return 1.0
```

```python
data['is_genuine']=data['is_genuine'].apply(convert)
```

```python
data['is_genuine'].describe
```

```python
from sklearn.model_selection import train_test_split
```

```python
data['is_genuine']
```

```python
data["diagonal"] = pd.to_numeric(data["diagonal"])
data["height_left"] = pd.to_numeric(data["height_left"])
data["height_right"] = pd.to_numeric(data["height_right"])
data["margin_low"] = pd.to_numeric(data["margin_low"])
data["margin_up"] = pd.to_numeric(data["margin_up"])
data["length"] = pd.to_numeric(data["length"])
```

```python
data.info()
```

```python
data['margin_low'] = data['margin_low'].fillna(value=data['margin_low'].mean())
```

```python
data['margin_low'].isna().sum()
```

```python

y = data['is_genuine']
data = data.drop('is_genuine',axis =1)
x = data
```

```python
x
```

```python
y
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=101)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train, y_train)
```

```python
pred = knn.predict(x_test)
```

```python
pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```

```python
# if we increase the neighbors
knn = KNeighborsClassifier(n_neighbors=15)
```

```python
knn.fit(x_train, y_train)
pred = knn.predict(x_test)
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```

hence we get a better accuracy on our model
